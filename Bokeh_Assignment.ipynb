**Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?**

We can customize the appearance of a plot, including the axes, title, and legend, by modifying various properties of the figure object.

```python
from bokeh.plotting import figure, show
from bokeh.models import Legend, LegendItem

p = figure()

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p.square(x, y, size=15, color="red", alpha=0.7, legend_label="Square Glyph")

# Customize axes
p.xaxis.axis_label = "X-axis"
p.yaxis.axis_label = "Y-axis"

# Customize title
p.title.text = "Customized Bokeh Plot"
p.title.text_color = "purple"
p.title.text_font_style = "bold"
p.title.text_font_size = "18pt"

# Customize legend
legend = Legend(items=[

    LegendItem(label="Square Glyph", renderers=[p.renderers[0]])
])
p.add_layout(legend, 'below')

# Show the plot
show(p)
```
